#### Q4. write a python programme to print the contents of a directory using os module. Search online for the function which does that.

```python
import os

# Specify directory path
directory_path = '/'

# List all files and directories in the specified path
contents = os.listdir(directory_path)

print(contents) # This will print the contents of the directory as a list

# If you want to print each item in the list on a new line, you can use a loop:
for i in contents:
    print(i)
```
