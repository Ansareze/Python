## **Chapter 2 - Variables and DataTypes**

### **Variable**
##### A variable is the name given to a memory location in a program.

### **Data Types**
##### Primarily there are the following types in python:
##### 1. Integers
##### 2. Floating Point Integers
##### 3. Strings
##### 4. Booleans
##### 5. None
##### Python automatically identifies the type of data for us.

```python
# variables = Container to store a  value
# keywords = reserved words in python
# identifiers = class/function/variable name

a = 30       # identifies a as class <int>
b = "John"   # identifies b as class <float>
c = 90.32    # identifies c as class <str>
```

### **Rules for choosing an identifier**
##### 1. A variable name can contain alphabets, digits and underscores.
##### 2. A variable name can only start with an alphabet and underscores.
##### 3. A variable name can't start with a digit.
##### 4. Noo white spaces is allowed to be used inside a variable name.
##### Example - one8, _seven, apple etc.

### Operators In Python
##### Following are some operators in python.
##### 1. Arithmetic operators : +, -, *, / etc.
##### 2. Assignment operators : +, +=, -= etc.
##### 3. Comparison operators : ==, >, >=, <, != etc. 
##### 4. Logical operators    : and, or, not

### Type() Function And Typecasting
##### type() function is used to find the data type of a given variable in python.

```python
a = 31
type(a)  # class <int>

b = "31"
type(b)  # class <str>
```

##### A number can be converted into a string and vice versa (if possible)
##### There are many functions to convert one data type into another.

```python
str(31)   # int to string
int("32") # string to integer
float(32) # integer to float
```

### Input () Function
##### This function allows the user to take input from the keyboard as a string.
##### output of 'input function' is always a string

```python
a = input("Enter Name")  # if user enters "hello world" then the a = "hello world"
```
